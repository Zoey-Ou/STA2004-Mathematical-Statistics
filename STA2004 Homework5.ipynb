```python
# STA2004 Programming assignment5
# Name: Ou Ziyi  Student ID:121090429


import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform
from scipy.stats import norm

# 2.1
n = 20
x = uniform.rvs(size = n)
epsilon = norm.rvs(loc = 0, scale = 0.1, size = n)
beta_0 = 1
beta_1 = 2
y = []
for i in range(n):
    yi = beta_0 + beta_1 * x[i] + epsilon[i]
    y.append(yi)
    
X_T = np.mat(x)
X = X_T.T
Y = np.mat(y).T

y_bar = sum(y)/n
x_bar = sum(x)/n
beta_1_hat = np.cov(x,y)[0][1]/np.var(x)
beta_0_hat = y_bar - beta_1_hat * x_bar
y_hat = beta_0_hat + beta_1_hat*x

plt.figure(figsize = (12,4))

plt.subplot(121)
plt.scatter(x,y)

plt.subplot(122)
plt.scatter(x,y)
plt.plot(x,y_hat)
```
